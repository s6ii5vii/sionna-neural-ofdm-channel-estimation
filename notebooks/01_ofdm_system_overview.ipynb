# OFDM system overview

This notebook explains the simplified pilot-observation model used by the reproducible experiments. Reusable simulation logic lives in `src/channel_estimation/`.

```python
import numpy as np
import matplotlib.pyplot as plt

from channel_estimation.ofdm import (
    generate_pilot_symbols,
    generate_rayleigh_channel,
    simulate_pilot_observations,
)
```

```python
rng = np.random.default_rng(42)
num_subcarriers = 64
pilots = generate_pilot_symbols(num_subcarriers)
channel = generate_rayleigh_channel((num_subcarriers,), rng)
received = simulate_pilot_observations(channel, pilots, snr_db=10, rng=rng)

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(np.abs(channel), label="True channel")
ax.plot(np.abs(received), "--", label="Noisy pilot observations")
ax.set(xlabel="Subcarrier index", ylabel="Magnitude", title="Simplified OFDM pilot model")
ax.grid(True, alpha=0.3)
ax.legend();
```

The current model uses independent Rayleigh coefficients and unit pilots. Future Sionna experiments should add standardized channel models and a complete resource-grid pipeline without changing the baseline API.
