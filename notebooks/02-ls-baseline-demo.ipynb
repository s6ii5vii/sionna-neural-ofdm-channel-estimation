# Least-squares baseline demo

This notebook demonstrates the implemented LS estimator and NMSE metric. The reproducible sweep is run from `experiments/baseline-ls/`.

```python
import numpy as np
import matplotlib.pyplot as plt

from channel_estimation.baselines import least_squares_estimate
from channel_estimation.metrics import normalized_mean_squared_error
from channel_estimation.ofdm import generate_pilot_symbols, generate_rayleigh_channel, simulate_pilot_observations
```

```python
rng = np.random.default_rng(42)
pilots = generate_pilot_symbols(64)
channel = generate_rayleigh_channel((64,), rng)
received = simulate_pilot_observations(channel, pilots, snr_db=10, rng=rng)
estimate = least_squares_estimate(received, pilots)

print(f"NMSE: {normalized_mean_squared_error(channel, estimate):.6f}")
```

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(np.abs(channel), label="True channel")
ax.plot(np.abs(estimate), "--", label="LS estimate")
ax.set(xlabel="Pilot index", ylabel="Magnitude", title="Least-squares channel estimate")
ax.grid(True, alpha=0.3)
ax.legend();
```
