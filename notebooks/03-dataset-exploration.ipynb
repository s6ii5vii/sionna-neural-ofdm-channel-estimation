# Dataset exploration

Inspect the committed sample dataset without embedding data-generation logic in the notebook.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

from channel_estimation.dataset import features_to_complex, inspect_dataset, load_npz_dataset
```

```python
dataset_path = Path("../data/channel-estimation-dataset.npz")
data = load_npz_dataset(dataset_path)
inspect_dataset(data)
```

```python
sample = 0
observation = features_to_complex(data["x_train"][sample])
channel = features_to_complex(data["y_train"][sample])

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(np.abs(channel), label="True channel")
ax.plot(np.abs(observation), "--", label="Noisy pilot observation")
ax.set(xlabel="Pilot index", ylabel="Magnitude", title="Committed sample dataset")
ax.grid(True, alpha=0.3)
ax.legend();
```
